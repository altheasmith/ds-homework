```python
import pandas as pd
```

```python
groups = pd.read_excel('Groups.xlsx', sheetname='World')
```

```python
groups = groups.fillna('')
```

```python
groups
```

```python
df = pd.read_excel('Groups.xlsx', sheetname='Pew Global')
```

```python
df.set_index('Country', inplace=True)
```

```python
df
```

```python
groups_by_country = pd.DataFrame(groups.groupby('Home Country')['Organization Name'].count())
```

```python
groups_by_country.rename(columns={'Organization Name': 'Terrorist Groups'}, inplace=True)
```

```python
groups_by_country.head()
```

```python
df = df.join(groups_by_country)
```

```python
df.head()
```

```python
df.head()
```

```python
colonies = pd.read_excel('Groups.xlsx', sheetname='British Colonies')
```

```python
colonies.set_index('Home Country', inplace=True)
```

```python
colonies.rename(columns={'Notes': 'BC Notes'}, inplace=True)
```

```python
df = df.join(colonies)
```

```python
df.head()
```

```python
df['BC Notes'].fillna('', inplace=True)
```

```python
t_info = df.dropna()
```

```python
len(t_info)
```

```python
df
```

```python
df['Terrorist Groups'].fillna(0, inplace=True)
```

```python
train_cols = list(df.columns)
```

```python
train_cols.remove('Terrorist Groups')
```

```python
train_cols.remove('BC Notes')
```

```python
train_cols
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
X = t_info[train_cols]
y = t_info['Terrorist Groups']
lm.fit(X, y)
```

```python
lm.coef_
```

```python
df_predict = df
```

```python
df_predict['Predict Terrorist Groups'] = lm.predict(df[train_cols])
```

```python
df_predict
```

```python
lm.coef_
```
